```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.model_selection import train_test_split
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.holtwinters import ExponentialSmoothing
```

```python
df=pd.read_csv("rainfall_2015_2023.csv")
df.head(10)
```

```python
df.info()
```

```python
df['date'] = pd.to_datetime(df['date'])
df.set_index('date', inplace=True)
```

```python
# Interpolate missing values
df['rainfall_mm'] = df['rainfall_mm'].interpolate(method='time')
```

```python
df.head(5)
```

```python
#extract month, year 
df['month'] = df.index.month
df['year'] = df.index.year
#df['rolling_7d'] = df['rainfall_mm'].rolling(window=7).mean()
#df['rolling_30d'] = df['rainfall_mm'].rolling(window=30).mean()
df["rolling_7d"] = df["rainfall_mm"].rolling(7).mean()
df["rolling_30d"] = df["rainfall_mm"].rolling(30).mean()
```

```python
df.head(5)
```

```python
df.rolling_30d.isna().sum()
```

```python
df.head(10)
```

```python
#  Plot 1: Daily Rainfall + 30-Day Trend 
plt.figure(figsize=(14, 6))
plt.plot(df.index, df["rainfall_mm"], alpha=0.3, label="Daily Rainfall")
plt.plot(df.index, df["rolling_30d"], color="red", linewidth=2,
         label="30-Day Rolling Average")
plt.title("Daily Rainfall Trend (20152023)")
plt.xlabel("Date"); plt.ylabel("Rainfall (mm)")
plt.legend()
plt.savefig("trend_daily.png")
#plt.close()
plt.show()
```

```python
# Plot 2: Seasonal Decomposition 
try:
    decomposition = seasonal_decompose(df["rainfall_mm"], model="additive", period=365)
    fig = decomposition.plot()
    fig.set_size_inches(12, 8)
    fig.savefig("seasonal_decomposition.png")
    #plt.close(fig)
    plt.show()
except Exception as e:
    print(" Decomposition Failed:", e)
```

#### Trend: Shows the overall long-term direction (mostly flat, as observed above).
#### Seasonal: Shows a highly consistent annual cycle, peaking in the middle of the year (May-July )follwed by September -October and dipping at the start/end of the year(December- February).
#### Residual: What is left after accounting for trend and seasonality (the unpredictable daily variability).
#### Noise :Daily rainfall has irregular short-term variations.

```python
# Plot 3: Monthly Rainfall Distribution 
plt.figure(figsize=(10, 6))
sns.boxplot(x=df["month"], y=df["rainfall_mm"])
plt.title("Monthly Rainfall Distribution")
plt.xlabel("Month"); plt.ylabel("Rainfall (mm)")
plt.savefig("monthly_boxplot.png")
#plt.close()
plt.show()
```

#### Some months show higher spread, indicating unreliable rainfall patterns.
#### Peak rainfall months show frequent extreme values.
#### The boxplot clearly shows that Months 4 through 8 have the highest average rainfall and the largest extremes (outliers), indicating the wet season.

```python
# Plot 4: Heatmap (Month vs Year)
monthly_totals = df.pivot_table(values="rainfall_mm", index="year",
                                columns="month", aggfunc="sum")

plt.figure(figsize=(12, 8))
sns.heatmap(monthly_totals, annot=True, fmt=".0f", cmap="Blues")
plt.title("Monthly Rainfall Heatmap (mm)")
plt.savefig("heatmap_monthly.png")
#plt.close()
plt.show()
```

#### The heatmap shows the total rainfall for each month/year combination. The darker blues consistently fall in the mid-year months, confirming the same seasonal pattern holds true across every year.

```python
# Convert daily to monthly rainfall totals
df_monthly = df["rainfall_mm"].resample("M").sum().to_frame(name="rainfall")

# Creating a lag features for ML model (112 months)
df_lags = df_monthly.copy()
for i in range(1, 13):
    df_lags[f"lag_{i}"] = df_lags["rainfall"].shift(i)

df_lags.dropna(inplace=True)

# Train-test split (80% train and 20% test)
train_size = int(len(df_lags) * 0.8)
train, test = df_lags.iloc[:train_size], df_lags.iloc[train_size:]
```

```python
# a function for model evaluation
def evaluate(y_true, y_pred, model_name):
    """Return RMSE, MAE, and R2 for any model."""
    return {
        "Model": model_name,
        "RMSE": np.sqrt(mean_squared_error(y_true, y_pred)),
        "MAE": mean_absolute_error(y_true, y_pred),
        "R2": r2_score(y_true, y_pred)
    }
```

#### Training Models 

```python
# Model 1 we using the ARIMA 
try:
    arima = ARIMA(train["rainfall"], order=(5, 1, 0)).fit()
    arima_pred = pd.Series(arima.forecast(len(test)), index=test.index)
    arima_results = evaluate(test["rainfall"], arima_pred, "ARIMA")
except Exception:
    arima_results = {"Model": "ARIMA", "RMSE": 1e9, "MAE": 1e9, "R2": -999}
```

```python
# Model 2 where we use the Random Forest
X_train = train.drop("rainfall", axis=1)
y_train = train["rainfall"]
X_test = test.drop("rainfall", axis=1)

rf = RandomForestRegressor(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)
rf_pred = pd.Series(rf.predict(X_test), index=test.index)
rf_results = evaluate(test["rainfall"], rf_pred, "Random Forest")
```

```python
# Model 3 in which we use the Holt-Winters/Exponential Smoothing
hw = ExponentialSmoothing(train["rainfall"],
                          seasonal="add",
                          seasonal_periods=12).fit()

hw_pred = pd.Series(hw.forecast(len(test)), index=test.index)
hw_results = evaluate(test["rainfall"], hw_pred, "Holt-Winters")
```

```python
# Comparing all the 3 models
results = pd.DataFrame([arima_results, rf_results, hw_results])
print("\n Model Performance Comparison:")
print(results)

# Select model with lowest RMSE
best_model = results.sort_values("RMSE").iloc[0]["Model"]
print(f"\n Best Model: {best_model}")
```

#### The Holt-Winters model was selected for the final forecast due to its superior performance ($R^2 \approx 0.99$) and its specific suitability for data with clear seasonality and no significant trend.

```python
# Making a 12 month Forecast
final_model = ExponentialSmoothing(df_monthly["rainfall"],
                                   seasonal="add",
                                   seasonal_periods=12).fit()

forecast_values = final_model.forecast(12)
forecast_dates = pd.date_range(df_monthly.index[-1] + pd.DateOffset(months=1),
                               periods=12, freq="M")

forecast_series = pd.Series(forecast_values, index=forecast_dates,
                            name="forecasted_rainfall")

# Save forecast
forecast_series.to_csv("forecast.csv")
```

```python
# A look at the Forecast Plot
plt.figure(figsize=(12, 6))
plt.plot(df_monthly.index, df_monthly["rainfall"], label="Historical")
plt.plot(forecast_series.index, forecast_series, marker="o", label="Forecast")
plt.title(f"12-Month Rainfall Forecast ({best_model})")
plt.xlabel("Date"); plt.ylabel("Rainfall (mm)")
plt.legend()
plt.savefig("final_forecast.png")
plt.show()
```

#### The final 12-month forecast, based on the Holt-Winters model, predicts future monthly rainfall based on the consistent historical seasonality.
#### Reliable Predictions: The forecast is highly reliable for predicting the timing and magnitude of the upcoming wet and dry seasons.
#### Actionable Insight: This forecast is most useful for operational planning, such as:

##### Agriculture: Timing planting and harvesting based on predicted wet season start/peak.
##### Water Management: Planning reservoir levels and conservation efforts ahead of the dry season.
##### Infrastructure: Preparing for periods of high flow (peak wet season) to prevent flooding.
